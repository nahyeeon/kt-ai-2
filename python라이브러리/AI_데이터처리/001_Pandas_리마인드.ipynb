### 실습 5 : Group by

* Q1. 객실 등급별 평균 나이

```python
titanic.groupby(by='Pclass', as_index=False)['Age'].mean()
```

* Q2. 승선 지역별 평균 운임

```python
titanic.groupby(by='Embarked', as_index=False)['Fare'].mean()
```
