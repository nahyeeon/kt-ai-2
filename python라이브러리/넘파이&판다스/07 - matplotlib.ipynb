## 1.환경준비

* 라이브러리 로딩

```python
import pandas as pd
import matplotlib.pyplot as plt
```

* 데이터 로딩

```python
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/retail_demand2.csv'
data = pd.read_csv(path, usecols = ['date','sales', 'comp_sales', 'tot_sales'])
data = data.loc[data['date']>= '2017-12-01'].reset_index(drop = True)
data['date'] = pd.to_datetime(data['date'])
data['tot_sales'] = data['tot_sales'] / 5
data.columns = ['date','item1', 'item2', 'item3']
data.head(10)
```

## 2.기본차트 그리기

1) 차트 그리기

* plt.plot(1차원 값)
    * x축 : 인덱스
    * y축 : 1차원 값
    * 라인차트를 그려줍니다.

```python
# 차트 그리기
plt.plot([2,5,3,1,2])

# 화면에 보여주기
plt.show()
```

* 데이터프레임으로 부터 데이터를 가져다 그릴 수 있습니다.

```python
plt.plot(data['item1'])
plt.show()
```

```python
plt.plot('item1', data = data)
plt.show()
```

2) x축과 y축의 값 지정하기  
* plt.plot(x축, y축)

* 값을 하나 넣으면, x축은 인덱스로 채워집니다. 
* 값을 두개 넣으면, 
    * x축 : 첫번째 값
    * y축 : 두번째 값

```python
# 날짜 축으로 sales에 대해 라인차트를 그려봅시다 파이썬은 이 코드를 권장 
plt.plot('date', 'item1', data = data)
plt.show()
```

```python
# 이렇게 넣어도 동일합니다.
plt.plot(data['date'], data['item1'])
plt.show()
```

* [추가] x 축의 값을 회전시키기

```python
plt.plot(data['date'], data['item1'])
plt.xticks(rotation=45)
plt.show()
```

## 4.추가기능

### (1) 그래프 크기 조정

    * default size는 6.4, 4.4

```python
plt.figure(figsize = (12,8))

plt.plot('date', 'item1', 'go--', data = data, label = 'apple')
plt.plot('date', 'item2', 'b*-', data = data, label = 'mango')
plt.plot('date', 'item3', 'rs-', data = data, label = 'banana')

plt.xlabel('day') 
plt.ylabel('sales')
plt.title('Daily Sales')

plt.legend()
plt.grid()
plt.show()
```

### (2) 수평선 수직선 추가

```python
plt.figure(figsize = (12,8))

plt.plot('date', 'item1', 'go--', data = data, label = 'apple')
plt.plot('date', 'item2', 'b*-', data = data, label = 'mango')
plt.plot('date', 'item3', 'rs-', data = data, label = 'banana')

# 각 아이템의 평균 선(수평선) 추가
plt.axhline(data['item1'].mean(), color = 'grey')
plt.axhline(data['item2'].mean(), color = 'grey')
plt.axhline(data['item3'].mean(), color = 'grey')

plt.xlabel('day') 
plt.ylabel('sales')
plt.title('Daily Sales')

plt.legend()
plt.grid()
plt.show()
```

### (3) 여러 그래프 나눠서 그리기

![](https://codetorial.net/matplotlib/_images/subplot_03.png)

* row : 고정된 행 수
* column : 고정된 열 수
* index : 순서

#### 1) 위, 아래로(3행 1열) 그래프 그리기.

```python
plt.figure(figsize = (12,8))

plt.subplot(3,1,1)
plt.plot('date', 'item1', 'go--', data = data, label = 'apple')
plt.legend()

plt.subplot(3,1,2)
plt.plot('date', 'item2', 'b*-', data = data, label = 'mango')
plt.legend()

plt.subplot(3,1,3)
plt.plot('date', 'item3', 'rs-', data = data, label = 'banana')
plt.legend()

plt.show()
```

#### 2) 옆으로 (1행 3열) 그래프 그리기.

```python
plt.figure(figsize = (12,8))

plt.subplot(1,3,1)
plt.plot('date', 'item1', 'go--', data = data, label = 'apple')
plt.legend()

plt.subplot(1,3,2)
plt.plot('date', 'item2', 'b*-', data = data, label = 'mango')
plt.legend()

plt.subplot(1,3,3)
plt.plot('date', 'item3', 'rs-', data = data, label = 'banana')
plt.legend()

plt.show()
```
