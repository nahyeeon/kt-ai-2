```python
path = 'https://raw.githubusercontent.com/JunhoYu/Data-Analysis/main/Carseats.csv'
data = pd.read_csv(path)
data.drop(['Income', 'Urban', 'Education'],axis = 1, inplace = True)
data.head()
```

```python
target = 'Sales'
```

### (1) 숫자형(수치형) X --> Y
* 모든 숫자형 X에 대해서 Y와 비교하여 차트를 그리고 수치화 하시오.

① 시각화 : scatter, jointplot  
② 수치화 : 상관분석

#### ① Advertising -> Sales

1) 시각화 : scatter(regplot), jointplot

```python
sns.scatterplot(x='Advertising', y =target, data = data) # 맨 왼쪽에 있는 0에 있는 직선은 광고비가 필요없는 해외 지점
plt.show()
```

```python
sns.regplot(x='Advertising', y =target, data = data)
plt.show()
```

```python
sns.jointplot(x='Advertising', y =target, data = data)
plt.show()
```

2) 수치화 : 상관분석

```python
# 상관계수와 p-value 
spst.pearsonr(data[target], data['Advertising'])
```

3) 시각화와 수치화 결과로부터 파악한 내용을 적어 봅시다.

### (2) 범주형 X --> Y

* 모든 범자형 X에 대해서 Y와 비교하여 차트를 그리고 수치화 하시오.

① 시각화 : 평균비교 barplot  
② 수치화 : t-test, anova

#### ① ShelveLoc-> Sales

1) 시각화 : 평균비교 barplot

```python
# Pclass(3 범주)
sns.barplot(x="ShelveLoc", y=target, data=data)
plt.grid()
plt.show()
```

```python
data['ShelveLoc'].unique()
```

2) 수치화 : t-test, anova

```python
E_1= data.loc[data.ShelveLoc=='Bad', 'Sales']
E_2= data.loc[data.ShelveLoc=='Good', 'Sales']
E_3= data.loc[data.ShelveLoc=='Medium', 'Sales']
spst.f_oneway(E_1,E_2,E_3)
```

3) 시각화와 수치화 결과로부터 파악한 내용을 적어 봅시다.

#### ② US-> Sales

1) 시각화 : 평균비교 barplot

```python
sns.barplot(data=data, x='US', y=target)
plt.grid()
plt.show()
```

```python
data['US'].unique()
```

2) 수치화 : t-test, anova

```python
male = data.loc[data['US']=='Yes', 'Sales']
female = data.loc[data['US']=='No','Sales']
spst.ttest_ind(male, female) 
```

3) 시각화와 수치화 결과로부터 파악한 내용을 적어 봅시다.

관련이 있다 𝐹 통계량 =
(집단 간 분산)/(집단 내 분산) = (전체 평균 − 각 집단 평균)/(각 집단의 평균 − 개별 값)
값이 대략 2~3 이상이면 차이가 있다고 판단합니다.

```python
temp = titanic.loc[titanic[feature].notnull()]

model = sm.Logit(temp[target], temp[feature])
result = model.fit()
print(result.pvalues)
```

### (2) 범주형 X --> 범주형 Y

* 교차표(pd.crosstab())
* 시각화( Stacked Bar, Mosaic)
* 가설검정(카이제곱검정)
* 관계 평가

#### ① Sex --> Survived

```python
feature='Sex'
```

1) 교차표(pd.crosstab())

```python
pd.crosstab(titanic[feature],titanic[target],normalize='index')
```

2) 시각화( Stacked Bar, Mosaic)

```python
temp=pd.crosstab(titanic[feature], titanic[target], normalize='index')
temp.plot.bar(stacked=True)
plt.axhline(1-titanic[target].mean(),color='r')
plt.show()
```

```python
mosaic(titanic, [feature,target])
plt.axhline(1- titanic[target].mean(), color = 'r')
plt.show()
```

3) 가설검정(카이제곱검정)

```python
table = pd.crosstab(titanic[target], titanic[feature])
result = spst.chi2_contingency(table)
print('카이제곱통계량', result[0])
print('p-value', result[1])
```
