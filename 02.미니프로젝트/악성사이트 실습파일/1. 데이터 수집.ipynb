```python
# <title> 태그 출력
print("* html :",soup.find('html'))

# <script> 태그 길이 출력
print("* html 길이 :", (soup.find('html').getText().count(' ')))
```

```python
# Feature(특징) 데이터를 추출하는 함수를 작성합니다.

def html_empty(soup):
    try:
        return(soup.find('html').getText().count(' '))
    except:
        return 0.0
```

```python
# 데이터 프레임의 html_code 컬럼에서 Feature(특징) 데이터를 추출합니다.

html_len = []
for index, row in df.iterrows():  # 행 하나씩 출력하는
    soup = BeautifulSoup(row.html_code, 'html.parser')
    html_len.append(html_empty(soup))
```

```python
# 추출한 Feature(특징) 데이터를 확인합니다.
df['html_empty'] = html_len
```

```python
df
```

```python
def spacebar_count(soup):
    try:
        Null_Count = soup.body.text.count(' ')
        return float(Null_Count)
    except:
        return 0.0
```

```python
white_space = []
for index, row in df.iterrows():
    soup = BeautifulSoup(row.html_code, 'html.parser')
    white_space.append(spacebar_count(soup))
```
